```python
na_sales = df['NA_Sales']
eu_sales = df['EU_Sales']
jp_sales = df['JP_Sales']
other_sales = df['Other_Sales']
global_sales = df['Global_Sales']
# Descriptive statistics
na_mean = na_sales.mean()
na_median = na_sales.median()
na_mode = na_sales.mode()[0]
na_std_dev = na_sales.std()

eu_mean = eu_sales.mean()
eu_median = eu_sales.median()
eu_mode = eu_sales.mode()[0]
eu_std_dev = eu_sales.std()

jp_mean = jp_sales.mean()
jp_median = jp_sales.median()
jp_mode = jp_sales.mode()[0]
jp_std_dev = jp_sales.std()

other_mean = other_sales.mean()
other_median = other_sales.median()
other_mode = other_sales.mode()[0]
other_std_dev = other_sales.std()

global_mean = global_sales.mean()
global_median = global_sales.median()
global_mode = global_sales.mode()[0]
global_std_dev = global_sales.std()

print(f"NA Sales Mean: {na_mean}")
print(f"NA Sales Median: {na_median}")
print(f"NA Sales Mode: {na_mode}")
print(f"NA Sales Standard Deviation: {na_std_dev}")

print(f"EU Sales Mean: {eu_mean}")
print(f"EU Sales Median: {eu_median}")
print(f"EU Sales Mode: {eu_mode}")
print(f"EU Sales Standard Deviation: {eu_std_dev}")

print(f"JP Sales Mean: {jp_mean}")
print(f"JP Sales Median: {jp_median}")
print(f"JP Sales Mode: {jp_mode}")
print(f"JP Sales Standard Deviation: {jp_std_dev}")

print(f"Other Sales Mean: {other_mean}")
print(f"Other Sales Median: {other_median}")
print(f"Other Sales Mode: {other_mode}")
print(f"Other Sales Standard Deviation: {other_std_dev}")

print(f"Global Sales Mean: {global_mean}")
print(f"Global Sales Median: {global_median}")
print(f"Global Sales Mode: {global_mode}")
print(f"Global Sales Standard Deviation: {global_std_dev}")
```

# **Interpretation of Box Plots:**
Box plots provide information about the distribution, spread, and presence of outliers in your data. Interpret the key components of a box plot:
The box represents the interquartile range (IQR), with the median indicated by the line inside the box.
Whiskers extend to the minimum and maximum values within a certain range (often 1.5 times the IQR).
Outliers are data points beyond the whiskers.

## **Bivariate Analysis**

Bivariate analysis is a statistical method examining how two different things are related. The bivariate analysis aims to determine if there is a statistical link between the two variables and, if so, how strong and in which direction that link is.


```python
# Scatter Plot between NA_Sales and EU_Sales
plt.figure(figsize=(8, 6))
sns.scatterplot(x='NA_Sales', y='EU_Sales',hue='Year',data=df, color='skyblue')
plt.title("Scatter Plot: NA_Sales vs. EU_Sales")
plt.xlabel("NA_Sales")
plt.ylabel("EU_Sales")
plt.show()

# Scatter Plot between Global_Sales and Year
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Global_Sales', y='Year',hue='Genre',data=df, color='skyblue')
plt.title("Scatter Plot: Global_Sales vs. Year")
plt.xlabel("Global_Sales")
plt.ylabel("Year")
plt.show()
```

### **Conclusion from above scatter plots :      **
 The scatter plot of **NA_Sales versus EU_Sales** reveals a positive linear relationship between video game sales in North America (NA) and the European Union (EU). The data points generally cluster around a diagonal line, indicating that games performing well in one region tend to perform well in the other. However, while there is a moderate positive correlation, there are notable exceptions in the form of outliers where certain games achieve exceptional success in one region but not the other. These outliers may reflect specific game preferences or marketing strategies. This analysis suggests that understanding regional differences and tailoring marketing and distribution strategies accordingly could be beneficial for video game publishers and developers seeking to maximize sales in both regions.


The scatter plot of **Global_Sales against Year** offers valuable insights into the trends of global video game sales over time. It shows a clear pattern of increasing global sales from earlier years, with occasional peaks and dips. While there isn't a perfectly linear relationship, the general trend suggests a positive correlation between video game sales and time, indicating growth in the industry. The presence of outliers, especially during certain years, may represent significant game releases or exceptional market conditions. This analysis implies that the video game industry has experienced overall growth, but further investigation into the factors driving sales during peak years and strategies for sustaining growth during low years would be beneficial. Time-based analysis, such as trend forecasting, can help game developers and publishers make informed decisions regarding release timing and market strategies to maximize global sales.

### **Conclusion from the grouped bar chart : **
The grouped barplot, featuring 'Year' on the x-axis, 'Global_Sales' on the y-axis, and 'Genre' as the hue, provides a comprehensive view of how video game sales have evolved over time within various genres. It showcases shifting genre preferences among gamers, highlighting stalwart genres like 'Action' and 'Sports' alongside more niche genres such as 'Puzzle' and 'Strategy.' Yearly fluctuations in sales hint at the industry's dynamic nature, with peaks and troughs likely influenced by blockbuster releases, technological advancements, or economic conditions. This analysis underscores the diversity in the industry, emphasizing the potential for strategic decisions in game development and marketing based on genre-specific trends and market opportunities.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('dvproject.csv')
genre_sales = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
stacked_data = genre_sales.stack().reset_index().rename(columns={0: 'Sales', 'level_1': 'Region'})
plt.figure(figsize=(12, 6))
plt.bar('Genre', 'Sales', data=stacked_data, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.xlabel('Genre')
plt.ylabel('Total Sales')
plt.title('Stacked Bar Chart: Regional Sales by Genre')
plt.legend(labels=['NA Sales', 'EU Sales', 'JP Sales', 'Other Sales'])
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()


```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(x='Genre', y='Global_Sales', data=df)
plt.xlabel('Genre')
plt.ylabel('Global Sales')
plt.title('Strip Plot: Global Sales by Genre')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## **Conclusion from above strip plot : **
The stripplot visually captures the distribution of global video game sales across different genres. It reveals several key insights into genre-specific sales trends. Notably, the 'Action' genre stands out with a wide distribution of sales, indicating both high- and low-performing titles within this category. 'Shooter' games exhibit a concentration of high sales figures, suggesting a strong market for this genre. Conversely, genres like 'Puzzle' and 'Strategy' tend to have lower sales figures and a narrower range, indicating a more specialized audience. The 'Role-Playing' genre showcases a broad range of sales, reflecting its versatility and enduring popularity. Overall, this analysis highlights the diverse landscape of video game sales across genres, emphasizing opportunities for developers and publishers to tailor their strategies to specific market segments and the need to consider genre preferences when planning game releases and marketing efforts.

## **Multivariate Analysis**

Multivariate analysis (MVA) involves evaluating multiple variables (more than two) to identify any possible association among them.

```python
columns_of_interest = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
correlation_data = df[columns_of_interest]
correlation_matrix = correlation_data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.xlabel('Sales Columns')
plt.ylabel('Sales Columns')
plt.show()
```

### **Conclusion from above heatmap : **
In multivariate analysis, a heatmap can be used to display the correlations between multiple variables, often in the form of a correlation matrix. Each cell in the matrix represents the correlation between two variables, and the heatmap color-codes these correlations for easy visual interpretation. This helps you understand how different variables are related to each other within a multivariate context.

The correlation heatmap offers a comprehensive view of the relationships between regional and global video game sales. Notably, it reveals strong positive correlations between North American and European sales, suggesting that games performing well in one of these regions tend to excel in the other. Conversely, the relatively weaker correlations with Japanese and Other region sales highlight distinct regional preferences and market dynamics. Of particular significance is the high positive correlation between global sales and North American and European sales, indicating that these regions exert a substantial influence on a game's global performance. The heatmap underscores the importance of understanding regional variations and underscores the need for tailored marketing and distribution strategies to accommodate diverse market dynamics.

```python
sales_data = df[['NA_Sales', 'EU_Sales', 'JP_Sales']]
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
x = sales_data['NA_Sales']
y = sales_data['EU_Sales']
z = sales_data['JP_Sales']
ax.scatter(x, y, z, c='b', marker='o')
ax.set_xlabel('NA Sales')
ax.set_ylabel('EU Sales')
ax.set_zlabel('JP Sales')
plt.title('3D Scatter Plot: NA Sales vs. EU Sales vs. JP Sales')
plt.show()
```

### **Conclusion from the above 3d scatter plot : **
The 3D scatter plot provides a dynamic visualization of video game sales across North America, Europe, and Japan. It reveals a complex interplay of sales patterns, demonstrating that while North American and European sales are often positively correlated, there are distinct variations within the Japanese market. The plot showcases the diversity of video game preferences, with some games achieving high sales across all three regions, while others experience regional disparities. The presence of clusters and outliers in the plot suggests the influence of unique factors or standout game titles in specific markets. This analysis highlights the importance of tailoring marketing strategies to regional nuances and understanding the multi-dimensional nature of global sales trends in the video game industry.
